```python
import pandas as pd
```

```python
df = pd.read_csv('OLAP_SWIGGY.csv')
```

```python
#Slice Operation
#Selecting Only rows where delivery status is delivered.
Slice_condition = df['DeliveryStatus'] == 'Delivered'
Sliced_data = df[Slice_condition]
Sliced_data  # Sliced_data.head(10)
```

```python
#Dice Operation
#Select only the rows where the 'DeliveryStatus' is 'Delivered' and the 'MenuItemName' contains the word 'Pizza'.
Dice_condition = (df['DeliveryStatus'] == 'Delivered') & (df['MenuItemName'].str.contains('Pizza'))
Diced_data = df[dice_condition]
Diced_data  #Diced_data.head(10)
```

```python
#Drill-Down Operation
#Breaking down the data based on the 'DeliveryStatus' and 'RestaurantName'.
Drill_down_data = df.groupby(['DeliveryStatus', 'RestaurantName']).count().reset_index()
Drill_down_data.head(10)
```

```python
#Roll-Up Operation
#Grouping the data by 'DateOfDelivery' to get the total counts for each delivery date.
roll_up_data = df.groupby(['DateOfDelivery']).count().reset_index()
roll_up_data.head(20)
```

```python
#Pivot Operation
#Show the total counts of each 'DeliveryStatus' for each 'RestaurantName'
pivot_data = df.pivot_table(index='RestaurantName', columns='DeliveryStatus', values='CustomerID', aggfunc='count', fill_value=0)
pivot_data.head(20)
```
